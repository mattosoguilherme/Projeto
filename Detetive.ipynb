```python
from time import sleep #importando apenas uma função da biblioteca time, para não deixar o codigo pesado
jogador=input('Digite o nome do personagem que você quer jogar:').strip().capitalize() #input para que o usuario insira o nome do personagem
personagem=jogador

if personagem!='Loise'and personagem!="Carlo" and personagem!="Nora"and personagem!="Frank" and personagem!="Lorenzo": #condição criada para mostrar mensagem que o jogador escreveu errado
  print('Você digitou o nome errado, este personagem não está na lista!')
elif personagem=='Loise'or personagem=="Carlo" or personagem=="Nora"and personagem!="Frank" or personagem!="Lorenzo": #condição elif criada para dar sequencia ao programa caso usuario coloque o nome certo
  print("\nAs investivações começaram...")
  sleep(2)#tempo de uma fala para outra
  print('\nSherlok chega a cena do crime para investigar o que ocorreu...')
  sleep(3)
  print('\nEle não encontra na conclusivo no local, nenhuma digital ou DNA do assassino...')
  sleep(5)
  print("\nFrustrado com a perícia, Sherlok resolve interrogar algumas pessoas que conheciam o Gaetano Lucchese!")
  sleep(5)
  print(f'\nSherlok:\n- Senhor(a) {personagem}, chamei você aqui na delegacia,  para fazer algumas perguntas bem simples, responda com sinceridade e poderá ir embora.')
  sleep(5)
  print(f'{personagem}:\n- Tudo bem, Senhor Holmes.')
  sleep(3)
  pergunta1=str(input(f'\nSherlok:\n- Você telefonou para a vítima?\n{personagem}:\n- ')).strip().capitalize()[0]#input que irá receber valor em str,com a primeira letra maiscula, as demais funções nessa linha são para descartar qualquer caracter que venha após do primeiro.
if pergunta1!="S"and pergunta1!="N":# condição criada para mostrar mensagem de erro, caso o usuario quebre as regras
  print("!ERRO!Você quebrou as regras!")
else:
  sleep(2)
  pergunta2=str(input(f'\nSherlok:\n- Você esteve no local do crime?\n{personagem}:\n- ')).strip().capitalize()[0]
if pergunta2!="S"and pergunta2!="N":
  print("!ERRO!Você quebrou as regras!")
else:  
  sleep(2)#tempo certo 2
  pergunta3=str(input(f'\nSherlok:\n- Você mora perto da vítima?\n{personagem}:\n- ')).strip().capitalize()[0]
if pergunta3!="S"and pergunta3!="N":
  print("!ERRO!Você quebrou as regras!")
else: 
  sleep(2)#tempo certo 2
  pergunta4=str(input(f'\nSherlok:\n- Você devia para a vítima?\n{personagem}:\n- ')).strip().capitalize()[0]
if pergunta4!="S"and pergunta4!="N":
  print("!ERRO!Você quebrou as regras!")
else:
  sleep(2)#tempo certo 2
  pergunta5=str(input(f'\nSherlok:\n- Você já trabalhou com a vítima?\n{personagem}:\n- ')).strip().capitalize()[0]
if pergunta5!="S"and pergunta5!="N":
  print("!ERRO!Você quebrou as regras!")
#condição criada para sobrepor o valor das variaves, invertendo-as de str para int
if pergunta1=="S": 
  pergunta1=1
elif pergunta1=="N":
  pergunta1=0

if pergunta2=="S":
  pergunta2=1
elif pergunta2=="N":
  pergunta2=0

if pergunta3=="S":
  pergunta3=1
elif pergunta3=="N":
  pergunta3=0

if pergunta4=="S":
  pergunta4=1
elif pergunta4=="N":
  pergunta4=0

if pergunta5=="S":
  pergunta5=1
elif pergunta5=="N":
  pergunta5=0

decisao= pergunta1+pergunta2+pergunta3+pergunta4+pergunta5 #variavel criada para somar os valores int de acordo com as respostas o usuario
#codições abaixo foram criada para mostrar finais diferentes de acordo numero de respostas positivas do usuario
if decisao <=1:#inocente
  sleep(2)
  print(f"\nSherlok:\n- Muito obrigado sr(a) {personagem}, pela cooperação. Você é inocente e está liberado!")
elif decisao==2:#suspeito
  sleep(2)
  print(f"\nSherlok:\n- Obrigado pela cooperação, mas você acaba de se enquadrar como supeiro do meu caso.\n Sr(a) {personagem} está probida de sair da cidade, caso saia teremos que prender você!")
elif decisao ==3 or decisao == 4:#cumplice
  sleep(2)
  print(f'\nSherlok:\n- Sr(a) {personagem}, você está sendo preso por ser cúmplice do assassinato de Gaetano Tommy Brown Lucchese, você o direito de permanercer calado...') 
elif decisao ==5:#culpado
  sleep(2)
  print(f'\nSherlok:\n- Sr(a) {personagem}, você é culpado pelo assassinato de Gaetano Tommy Brown Lucchese,\n a pena por esse crime é de 6 a 20 anos de prisão, prepare-se para apodrecer na cadeia!'
  )
```
